```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the diabetes prediction dataset (replace 'diabetes_prediction_dataset.csv' with the actual file path)
data = pd.read_csv('diabetes_prediction_dataset.csv')

# Scatter plot with regression line to show the relationship between age and blood glucose level
plt.figure(figsize=(10, 6))
sns.regplot(x="age", y="blood_glucose_level", data=data, scatter_kws={'s': 10}, line_kws={'color': 'red'})
plt.title("Relationship Between Age and Blood Glucose Level")
plt.xlabel("Age")
plt.ylabel("Blood Glucose Level")
plt.show()
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the diabetes prediction dataset (replace 'diabetes_prediction_dataset.csv' with the actual file path)
data = pd.read_csv('diabetes_prediction_dataset.csv')

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x="age", y="blood_glucose_level", data=data, scatter_kws={'s': 10}, line_kws={'color': 'red'})
plt.title("Relationship Between Age and Blood Glucose Level")
plt.xlabel("Age")
plt.ylabel("Blood Glucose Level")

# Display the plot in Streamlit
st.pyplot()

# Display a subset of the data
st.header("Subset of the Data")
subset_data = data.head(10)  # Displaying the first 10 rows as an example
st.dataframe(subset_data)

```
